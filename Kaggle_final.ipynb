```python
## 7) Missing value imputation
## 1st method -
## 1a)Categorical variables - replace with mode
## 1b)Numerical variables - replace with mean(if normal distribution:Age), with median (if skewed:Fare)

## 2nd method -  Above method invlolve biasness, so another way :
## 2a)group the data by some variables that have no NA's & for each subset compute with median
## 2b)build linear regression model that will predict NA's using features that have no NA's

## by using 2a)
## if 1 or 2 missing NA, use 1st method
## if more NA, use 2nd method - find out which variables have strong relation - using correlation 

merge.isnull().sum()
# Age - 263 : 2nd method
# Fare - 1 : 1st method
# Cabin - 1014 : 2nd method
# Emabrked - 2 : 1st method
# Survived - 418 : ignore as these we have to predict
```

```python
## Embarked : As categorical variable - replace with mode
## Fare : As skewed distribution - replace with median
#merge.dtypes #Age-float64, Fare-float64, Embarked-object
merge['Embarked'].fillna(value='S',inplace=True)  # the most occurence of S, Embarked-object
merge['Fare'].fillna(value=merge['Fare'].median(),inplace=True)  
```

```python
## Function to tune hyperparameters
def tune_hyperparameters(model, params):
    from sklearn.model_selection import GridSearchCV
    global best_params, best_score
    ##construct grid search object with 10 fold cross validation
    grid=GridSearchCV(model,params,verbose=0,cv=10,scoring='accuracy',n_jobs=-1) 
    grid.fit(X_train,y_train)  ##fit using grid search
    best_params,best_score=grid.best_params_,np.round(grid.best_score_*100,2)
    return best_params,best_score

tune_hyperparameters(lr,params=lr_params)
lr_best_params, lr_best_score = best_params, best_score
display('LR Best Score :', lr_best_score, '& best parameters', lr_best_params)
```

```python
tune_hyperparameters(gbc,params=gbc_params)
gbc_best_params, gbc_best_score = best_params, best_score

tune_hyperparameters(svc,params=svc_params)
svc_best_params, svc_best_score = best_params, best_score

tune_hyperparameters(dt,params=dt_params)
dt_best_params, dt_best_score = best_params, best_score

tune_hyperparameters(rf,params=rf_params)
rf_best_params, rf_best_score = best_params, best_score

tune_hyperparameters(knn,params=knn_params)
knn_best_params, knn_best_score = best_params, best_score

tune_hyperparameters(abc,params=abc_params)
abc_best_params, abc_best_score = best_params, best_score

tune_hyperparameters(etc,params=etc_params)
etc_best_params, etc_best_score = best_params, best_score

#tune_hyperparameters(xgbc,params=xgbc_params)
#xgbc_best_params, xgbc_best_score = best_params, best_score

tune_hyperparameters(bc,params=bc_params)
bc_best_params, bc_best_score = best_params, best_score
```

```python
##11) Model Building and Evaluation
## 3) Retrain & Predict using optimized hyper parameters : we will retrain the classifiers using their best hyperparameters
                                              ## & then use cross validation to calculate the accuracy of the trained model
    
## Instantiate the models with optimised hyperparameters
rf = RandomForestClassifier(**rf_best_params)
gbc = GradientBoostingClassifier(**gbc_best_params)
svc = SVC(**svc_best_params)
knn = KNeighborsClassifier(**knn_best_params)
etc = ExtraTreesClassifier(**etc_best_params)
lr = LogisticRegression(**lr_best_params)
dt = DecisionTreeClassifier(**dt_best_params)
abc = AdaBoostClassifier(**abc_best_params)
bc = BaggingClassifier(**bc_best_params)

## Train all the models with optimised hyperparameters
models = {'RF':rf, 'GBC':gbc, 'SVC':svc, 'KNN':knn, 'ETC':etc, 'LR':lr, 'DT':dt, 'ABC':abc, 'BC':bc}
score=[]
for x, (keys, items) in enumerate (models.items()):
    from sklearn.model_selection import cross_val_score
    items.fit(X_train,y_train)
    cross_val_score = np.round(cross_val_score(items, X_train, y_train, cv=10, scoring='accuracy')*100,2)
    score.append(cross_val_score.mean())
    display(keys, cross_val_score.mean())     
    
## Here, we managed to reproduce the same score that we achieved only after tuning hyperparameters -
                                                             ## RF & GBC have the highest accuracy after tuning hyper parameters
                                                             ## Hence, will use RF & GBC for further study
## Now, let's predict using these trained models, we will have the best test accuracy possible        
```

```python
##12) More Evaluation Metrics:
##So fare, the accuracy score we have used to evaluate our classifiers is not enough as accuracy score doesn't tell exactly- 
##-which class (positive or negative)is being wrongly classified by our classifier in case of low accuracy score

##For imbalanced classification problem, accuracy score is not the best metric to chose between different classifier
##We will use these methods(confusion matrix, ROC & AUC score) for both random forest(rf) & gradient boosting classifier(gbc) 
##Finally compare rf & gbc using these calculated metrics to see which classifier excels the other

##12.1) Confusion Matrix
##12.2) ROC Curve & AUC score

##Confusion Matrix

##Confusion Matrix shows the number of correct classifications along with misclassifications when a classifier make predictions- 
##-for each class(poisitve or negative). The diagonal elements are correct classifications while off diagonal elements are -
##-missclassifications. Terms - 
#a)True Positives(TP)-Cases in which predicted 1(yes), actual is also 1(yes)
#b)True Negatives(TN)-Cases in which predicted 0(no), actual is also 0(no)
#c)False Positives(FP)-Cases in which predicted 1(yes), actual is 0(no) : Also known as Type1 error
#d)False Negatives(FN)-Cases in which predicted 0(no), actual is 1(yes) : Also known as Type2 error

from sklearn.model_selection import cross_val_predict
rf_predicted = cross_val_predict(rf, X_train, y_train, cv=10)
rf_confusion_matrix=pd.crosstab(y_train, rf_predicted, rownames = ['Actual'], colnames = ['Predicted/Classified'], margins = True)
display('RF Confusion Matrix:', rf_confusion_matrix)

gbc_predicted = cross_val_predict(gbc, X_train, y_train, cv=10)
gbc_confusion_matrix=pd.crosstab(y_train, gbc_predicted, rownames = ['Actual'], colnames = ['Predicted/Classified'], margins = True)
display('GBC Confusion Matrix:', gbc_confusion_matrix)

##The first row of confusion matrix(error matrix) is about negative class(coz of 0, hence not survived)
##The 2nd row of confusion matrix(error matrix) is about positive class(coz of 1, hence survived)

     TP      TN        FP       FN
RF   245     504       45       97
GBC  257     492       57       85
## RF makes exactly same correct predictions (TP+TN) as gbc- 749
## Hence, rf & gbc have exactly same accuracy score
```

```python
## ROC Curve & AUC Score
##--ROC Curve--
# ROC (Receiver Operating Characteristic Curve) is a plot of TPR against FPR of a classifier
# It shows trade off between sensitivity & specificity (any increase in sensitivity will be accompained by a decrease in specificity)
##--AUC Score (Area under the ROC curve)--
# It is corresponding score to the AUC curve, simply computed by measuring the area under ROC curve, which is called AUC
# will plot both - ROC curve & AUC score together for rf & gbc

from sklearn.metrics import roc_curve, roc_auc_score
rf_probability = rf.predict_proba(X_train)[:,1]
plt.figure(figsize=(18,5))
rf_false_positive_rate, rf_true_positive_rate, rf_threshold=roc_curve(y_train,rf_probability)
rf_auc_score=roc_auc_score(y_train,rf_probability)
plt.plot(rf_false_positive_rate, rf_true_positive_rate, label = 'ROC CURVE, AREA' + str(rf_auc_score))
plt.plot([0,1],[0,1],'red',lw=3.7)
plt.xlabel('False Positive Rate (1-Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.axis([0,1,0,1])
plt.legend(loc=4)
plt.title('RF ROC curve with AUC score')
plt.show()

gbc_probability = gbc.predict_proba(X_train)[:,1]
plt.figure(figsize=(18,5))
gbc_false_positive_rate, gbc_true_positive_rate, gbc_threshold=roc_curve(y_train,gbc_probability)
gbc_auc_score=roc_auc_score(y_train,gbc_probability)
plt.plot(gbc_false_positive_rate, gbc_true_positive_rate, label = 'ROC CURVE, AREA' + str(gbc_auc_score))
plt.plot([0,1],[0,1],'red',lw=3.7)
plt.xlabel('False Positive Rate (1-Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.axis([0,1,0,1])
plt.legend(loc=4)
plt.title('GBC ROC curve with AUC score')
plt.show()

## The distance between blue & red line of RF is greater than the distance between blue & red line of GBC
## Hence, RF is better at discriminating between positivies and negatives then GBC
## Also, RF AUC score is approx 92% which is greater than GBC AUC score (approx 90%)
## The higher the area, the further the classifier is off the red diagonal line & vice-versa
## Since RF has more area under ROC curve than GBC, RF is more accurate
```
